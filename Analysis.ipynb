#### Generate plot of AuNPs on glass from .csv file

```python
from pyuvvis import TimeSpectra

SAVEKWDS = dict(dpi=200) #Dots per in

ts_full = TimeSpectra.from_csv('NPSAM_cropped.csv', 
                                 index_col=0,
                                 specunit='nm',
                                 name='Very nice sam',
                                 header_datetime='%Y-%m-%d %H:%M:%S')
ts_full.reference=0
ts_full.iunit = 'a'
ts_full.varunit='m'
ts = ts_full.nearby[:, 1.0:5.0] #Cut time range
```

```python
from pyuvvis.plotting import areaplot
areaplot(ts);
```

#### Define a function that computes a cumulative sum of the difference or absolute difference in the area curves.

```python
def sumdiff(array, absolute=False, cumsum=False):
    """ Sum of the differences of an array (usually array of areas).  If absolute, absolute difference is used.
    If cumsum, the cumulative sum is returned.
    """
    sdiff = []
    # Compute the forward difference up to i=final -1
    for idx, value in enumerate(array):
        if idx < len(array)-1:
            new = array[idx+1]
            out = new-value            
            if absolute:
                out = abs(out)        
            sdiff.append(out)

    if cumsum:
        sdiff = list(np.cumsum(sdiff)) 
    sdiff.append(sdiff[-1]) #REPEAST LAST VALUE
    return sdiff

DIFF = area.apply(sumdiff, axis=1)
ABSDIFF =  area.apply(sumdiff, axis=1, absolute=True)
CUMDIFF = area.apply(sumdiff, axis=1, cumsum=True)
ABSCUMDIFF = area.apply(sumdiff, axis=1, absolute=True, cumsum=True)
```

```python
DIFF
```

```python
areaplot(ts, title='Area vs Time')


f, (ax1, ax2) = plt.subplots(1,2 )

for OBJ in [DIFF, ABSDIFF, CUMDIFF, ABSCUMDIFF]:
    OBJ.varunit='m'

DIFF.transpose().plot(ax=ax1, color='r', padding=None)
ABSDIFF.transpose().plot(ax=ax1, color='g', padding=None)
ax1.legend(['$\Delta$ Area', '$|\Delta$ Area|'],  loc='upper left')

CUMDIFF.transpose().plot(ax=ax2, color='r', ls='--', padding=None)
ABSCUMDIFF.transpose().plot(ax=ax2, color='g', ls='--', padding=None)
ax2.scatter(ABSCUMDIFF.columns[-1], ABSCUMDIFF.values[0][-1], s=100, color='g')
ax2.scatter(CUMDIFF.columns[-1], CUMDIFF.values[0][-1], s=100, color='r')
ax2.set_title('Cumulative sum')
ax2.set_ylabel('')
ax2.legend(['$\sum\Delta$ Area', '$\sum|\Delta$ Area|'], loc='upper left');
```
